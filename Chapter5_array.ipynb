```java
//array algorithm
//5.3.5 bubble sort
double[] scores = {99.5,55.7,55.9,66.3,84.1,22.4};
for (int i=0;i<scores.length;i++){
    for(int j=0;j<i;j++){
        if(scores[j]<scores[j+1]){
            double temp;
            temp = scores[j];
            scores[j]=scores[j+1];
        }
    }
}
for(int i=0;i<scores.length;i++){
    System.out.println(scores[i]);
}

//5.3.6 quick sort
//选择基准：从数组中选择一个元素作为基准值。
//分区操作：
//将数组中所有比基准值小的元素移动到基准值的左侧，将所有比基准值大的元素移动到基准值的右侧。
//递归排序：
//对左侧和右侧的子数组递归进行快速排序。
// 快速排序的实现
public class QuickSort {

    public static void main(String[] args) {
        int[] array = {10, 7, 8, 9, 1, 5};
        System.out.println("Original array:");
        printArray(array);
        
        quickSort(array, 0, array.length - 1);
        
        System.out.println("Sorted array:");
        printArray(array);
    }

    // 快速排序方法
    static void quickSort(int[] array, int low, int high) {
        if (low < high) {
            // 找到基准索引
            int pivotIndex = partition(array, low, high);
            // 递归排序左右部分
            quickSort(array, low, pivotIndex - 1);
            quickSort(array, pivotIndex + 1, high);
        }
    }

    // 分区方法
    static int partition(int[] array, int low, int high) {
        // 选择最后一个元素作为基准
        int pivot = array[high];
        int i = (low - 1); // 最小元素的索引

        for (int j = low; j < high; j++) {
            // 如果当前元素小于或等于基准
            if (array[j] <= pivot) {
                i++;
                // 交换 array[i] 和 array[j]
                swap(array, i, j);
            }
        }
        // 交换 array[i + 1] 和 pivot（最后一个元素）
        swap(array, i + 1, high);
        return i + 1; // 返回基准索引
    }

    // 交换方法
    static void swap(int[] array, int i, int j) {
        int temp = array[i];
        array[i] = array[j];
        array[j] = temp;
    }

    // 打印数组
    static void printArray(int[] array) {
        for (int num : array) {
            System.out.print(num + " ");
        }
        System.out.println();
    }
}

// 运行主方法
QuickSort.main(null);



//5.3.7 array copy
int[] array1 = {2, 3, 5, 7, 11}; // 初始化第一个数组
int[] array2 = array1; // array2 是 array1 的引用

// 输出 array1 的内容
System.out.print("遍历 array1: ");
for (int i = 0; i < array1.length; i++) {
    System.out.print(array1[i] + ",");
}
System.out.println(); // 换行

// 输出 array2 的内容
System.out.print("遍历 array2: ");
for (int i = 0; i < array2.length; i++) {
    System.out.print(array2[i] + ",");
}
System.out.println(); // 换行

// 修改 array2 的第一个元素
array2[0] = 0; // 这里的修改会影响到 array1

System.out.println("array2[0] 修改为 0 之后：");

// 输出 array1 的内容
System.out.print("遍历 array1: ");
for (int i = 0; i < array1.length; i++) {
    System.out.print(array1[i] + ",");
}
System.out.println(); // 换行

// 输出 array2 的内容
System.out.print("遍历 array2: ");
for (int i = 0; i < array2.length; i++) {
    System.out.print(array2[i] + ",");
}
System.out.println(); // 换行
//array1和array2属于数组类型的引用变量，当通过“=”赋值时，array1将地址赋值给了array2，因此二者引用（指向）了同一个数组对象的内存空间，所以通过一个引用修改该对象的成员，会影响另外一个引用




```
